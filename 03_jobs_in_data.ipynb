## Top 5 Categorias

```python
top_five = df['job_category'].value_counts()[:5]
print(type(top_five))
```

```python
top_five
```

```python
top_five.plot(
    kind='pie', 
    autopct='%1.1f%%'
)
plt.title('top 5')
figure = plt.gcf()
figure.set_size_inches(4, 4)
plt.show()
```

```python
top_five = df['job_title'].value_counts()[:-5]
print(type(top_five))
```
